```python
import pandas as pd
import plotly.express as px

df = pd.read_csv('https://raw.githubusercontent.com/Liqgy/coffeeforsales/main/CoffeeShopSales.csv')
df.head()
```

```python
df.drop(columns=['store_id','product_id'],axis=1,inplace = True)
df
```

```python
df['transaction_time'] = pd.to_datetime(df['transaction_time'])

# สร้างคอลัมน์ใหม่ hour_of_day เพื่อเก็บชั่วโมงของเวลาการทำธุรกรรม
df['hour_of_day'] = df['transaction_time'].dt.hour
df
```

```python
selectedc = df[['transaction_time', 'unit_price', 'product_category','hour_of_day']]
selectedc
```

```python
rankk = selectedc['product_category'].value_counts()
rankk
```

```python
filterr = selectedc[(selectedc.product_category=="Coffee")|(selectedc.product_category=="Tea")]
filterr
```

```python
def categorize_time(hour):
    if 7 <= hour < 12:
        return "Morning"
    elif 12 <= hour < 17:
        return "Lunch"
    else:
        return "Evening"

filterr['transaction_time'] = pd.to_datetime(filterr['transaction_time'])

# สร้างคอลัมน์ใหม่ time_category และใช้ฟังก์ชั่น categorize_time เพื่อแบ่งกลุ่มช่วงเวลา
filterr['time_category'] = filterr['transaction_time'].dt.hour.apply(categorize_time)

filterr
```

```python
filterr['frequency'] = filterr.groupby('hour_of_day')['hour_of_day'].transform('count')
filterr
```

```python
fig = px.scatter(filterr,
                x='product_category',
                y='hour_of_day',
                color='unit_price',
                size='frequency',
                hover_name='time_category',
                category_orders={'time_category': ['Morning', 'Lunch', 'Evening']})
fig.show()
print(df.head())
```
